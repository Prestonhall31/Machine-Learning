## Task 3: Create new feature(s)

```python
### Store to my_dataset for easy export below.
my_dataset = data_dict
```

### Custom Feature Creation

- I think that there could be correlation between how much contact between poi and non-poi and their shared receipts can provide us with more insight into whether this person could potentially be a poi. 

- I also want to look at their total compensation package, total_payments plus total_stock_value to see if there could be some correlation there. 

I will create two new features based on these observations. 

```python
for name in my_dataset:
    
    ### Direct POI involvement factor
    try:
        my_dataset[name]['from_poi_with_shared_receipt_percentage'] = my_dataset[name]['from_poi_to_this_person'] / float(my_dataset[name]['shared_receipt_with_poi'])
    except:
        my_dataset[name]['from_poi_with_shared_receipt_percentage'] = np.nan
        
    
    ### Total compensation
    try:
        my_dataset[name]['total_compensation'] = my_dataset[name]['total_payments'] + float(my_dataset[name]['total_stock_value'])
    except:
        my_dataset[name]['total_compensation'] = np.nan
```

```python
### Extract features and labels from dataset for local testing
data = featureFormat(my_dataset, features_list, sort_keys = True)
labels, features = targetFeatureSplit(data)
```

## Task 4: Try a varity of classifiers
